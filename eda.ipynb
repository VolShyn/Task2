
Our data can be categorised into 2 groups:

**Integers**:

* Are in $[0, 499]$ interval
* Having same standard deviation $144$
* Almost the same quartiles $125$

Discrete values like these can be defined as some category; nonetheless, let's stick with numerical representation.

**Floats**:

*   $[0,14.999]$ interval
*   same std
*  same quartiles

By looking at the box-plot, we can observe that our data have the same distribution (Uniform);

Though columns `6`,`7`,`8` are different;

The `Target` variable has an interval $[0,100]$, which might be a percentage. Yet, It is still a float value, so we're solving the regression problem.

Additionally, we need to normalize our data because the scale differs greatly.

To prove that our data have uniform distribution, let's plot some of the columns:

```python
# 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(data=train['1'], ax=ax1)
sns.histplot(data=train['31'], ax=ax2)
sns.histplot(data=train['target'], ax=ax3)
```

Yep, they are uniformly distributed;

Let's have a closer look at columns `6`,`7`,`8`;

```python
# 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(data=train['6'], ax=ax1)
ax1.axvline(0, color='r', ls=':')
ax1.text(0, 0.99, 'median', color='r', ha='right', va='top', rotation=90,
        transform=ax1.get_xaxis_transform())
sns.histplot(data=train['7'], ax=ax2)
sns.histplot(data=train['8'], ax=ax3)
```

Okay, the `7`-th column is uniformly distributed normalized data. As it was already said, we need to normalize our data; Double normalization is not the best solution, so we'll just keep it like that;

Column `6` have bimodal distribution;

Column `8` is definitely a categorical data;

let's additionally explore correlation coefficient's

```python
corr = train.corr()['target'].sort_values(ascending=False)
print(corr[:11])  # 11 because target itself will be included

plt.figure(figsize=(8, 6))
sns.heatmap(train.corr(),
            cmap='coolwarm',
            center=0)
plt.tight_layout()
plt.show()
```

No linear relationship at all, EXCEPT column `6` and `8`

We'll build scatterplot to analyse it more precisely:

```python
sns.scatterplot(x=train["6"], y=train["8"])
```

It's a good news, actually.

We don't want to keep categorical data in case of regression, especially when it doesn't add any additional information to our model because it's just linearly connected with the `6`.

```python
train.drop(columns=['8'])
```

Last but least - we need to normalise our data.

For normalisation we'll take the `min-max` approach, as It is generally suitable here, because:

* We will keep our data in the same interval $[0,1]$ as `7`.
* It preserves the uniform's distribution properties (unlike Z-score, which assumes normal distribution)
* `min-max` works just fine If we don't have outliers (and we don't have them)  

`target` will not be normalised, because of the interpretability of the final results.
